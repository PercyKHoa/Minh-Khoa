```python
import yfinance as yf

ticker = 'BTC-USD'
df = yf.download(ticker)
df
```

```python
df = df.loc['2020-01-01':].copy()
```

```python
df['change_tomorrow'] = df['Adj Close'].pct_change(-1)
df.change_tomorrow = df.change_tomorrow * -1
df.change_tomorrow = df.change_tomorrow * 100
```

```python
df = df.dropna().copy()
df
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(max_depth=20,random_state=42)
```

```python
y = df.change_tomorrow
X = df.drop(columns='change_tomorrow')
```

```python
X
```

```python
y = df.change_tomorrow.dropna()
```

```python
y
```

```python
model.fit(X,y)
```

```python
model.score(X, y)
```

```python
# Use the last row of data for today's prediction
explanatory_today = X.iloc[[-1], :]

# Predict the percentage change for tomorrow
predicted_change_tomorrow = model.predict(explanatory_today)[0]
print(f"Predicted change for tomorrow: {predicted_change_tomorrow:.2f}%")

# Calculate the predicted price for tomorrow
current_price = df['Adj Close'].iloc[-1]
predicted_price_tomorrow = current_price * (1 + predicted_change_tomorrow / 100)
print(f"Predicted price for tomorrow: {predicted_price_tomorrow:.2f}")
```

```python
#Parametrize & optimize the investment strategy
```

```python
from backtesting import Strategy
```

```python
class Regression(Strategy):
    limit_buy = 1
    limit_sell = -5
    
    def init(self):
        self.model = model
        self.already_bought = False

    def next(self):
        explanatory_today = self.data.df.iloc[[-1], :]
        forecast_tomorrow = self.model.predict(explanatory_today)[0]
        
        if forecast_tomorrow > self.limit_buy and self.already_bought == False:
            self.buy()
            self.already_bought = True
        elif forecast_tomorrow < self.limit_sell and self.already_bought == True:
            self.sell()
            self.already_bought = False
        else:
            pass
```

```python
from backtesting import Backtest
```

```python
bt = Backtest(
    X, Regression, cash=100000,
    commission=.002, exclusive_orders=True
)
```
